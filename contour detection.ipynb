Import modules: 

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Function to display images in Jupyter Notebook

def display_image(img, title="Image", cmap_type=None):
    plt.imshow(img, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')  # Hides the axis values
    plt.show()
```

Read and resize image:

```python
image = cv2.imread('test1.jpg')
image = cv2.resize(image, (350, 350))
```

## Original Image

```python
# convert from BGR to RGB for proper color display with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# display the original image
display_image(image_rgb, "Original Image")
```

## GrayScale Image

```python
# convert to grayscale
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# display the grayscale image
display_image(img_gray, "Grayscale Image", cmap_type='gray')
```

## Applying Binary Threshold

```python
# apply binary threshold
ret, img_thresh = cv2.threshold(img_gray, 50, 255, cv2.THRESH_BINARY)

# display the binary image
display_image(img_thresh, "Binary Image", cmap_type='gray')
```

## Detection of Contours

```python
# detect contours on binary image
contours, _ = cv2.findContours(img_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

# draw contours on the original image (RGB format for display)
img_copy = image_rgb.copy()
cv2.drawContours(img_copy, contours, -1, (0, 255, 0), 2, lineType=cv2.LINE_AA)
```

## Final Display of the Image

```python
display_image(img_copy, "Contour Image")
```
